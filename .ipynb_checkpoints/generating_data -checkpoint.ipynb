## Generating Data 
### Plotting a simple line graph

```python
import matplotlib.pyplot as plt
```

1. `subplots()` can generate one or more plots in the same figure.
2. `fig` represents the entire figure.
3. `ax` represens the a single plot inside `fig`.
4. `plot()` try to plot data in a meaningful way which is `squares` in this case.
5. `plt.show()` opens Matplotlib's viewer and display plots.

```python
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares)

plt.show()
```

### Changing the Label Type and Line Thickness 
1. Thicker line.
2. Compare **Square Number1** and **Square Number2** to understand what `tick_params()` can achive.

```python
squares = [1, 4, 9, 16, 25]
fig, ax1 = plt.subplots()
fig, ax2 = plt.subplots()

ax1.plot(squares, linewidth = 3)

ax1.set_title("Square Numbers1", fontsize = 24)
ax1.set_xlabel("Value", fontsize = 14)
ax1.set_ylabel("Square of Value", fontsize = 14)

ax1.tick_params(axis='both', labelsize=14)


ax2.plot(squares, linewidth = 3)

ax2.set_title("Square Numbers2", fontsize = 24)
ax2.set_xlabel("Value", fontsize = 14)
ax2.set_ylabel("Square of Value", fontsize = 14)

plt.show()
```

### Using Built-in Styles

```python
stylelist = plt.style.available
print(stylelist)
print(len(stylelist))
```

1. There are 26 built-in styles in **Matplotlib**.
2. Choose 3 different styles to plot.
3. If *fivethirtyeight* has been choosed, then the title of this plot is **Square Numbers of fivethirtyeight**.

```python
from random import randint
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

used_index = []
for i in range(3):
    styleindex = randint(0, len(stylelist) - 1)
    while styleindex in used_index:
        styleindex = randint(0, len(stylelist) - 1)
    
    used_index.append(styleindex)
    plt.style.use(stylelist[styleindex])
    fig, ax = plt.subplots()

    ax.plot(input_values, squares, linewidth = 3)
    
    ax.set_title("Square Numbers of " + stylelist[styleindex], fontsize = 24)
    ax.set_xlabel("Value", fontsize = 14)
    ax.set_ylabel("Square of Value", fontsize = 14)

    ax.tick_params(axis='both', labelsize=14)
plt.show()
```

### Plotting and Styling Individual Points with scatter()
1. In order to emphasize individual points, for example, plot small values in one color and large values in another.
2. Use `scatter()` to plot single point.
3. Pass `(x, y)` to `scatter()` to plot.
4. `s` in `sactter()` is the size of point.
5. `sactter()` also a `c` argument for passing color of points. This argument can be passed through color name inside quotation marks and also RGB values. To define color in RGB values, pass a tuple with three decimal values, each for red, green and black, using values between 0 to 1. The larger the value is, the brighter the color is.

```python
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2, 4, s = 200)

ax.set_title("Single Point", fontsize = 24)
ax.set_xlabel("x", fontsize = 14)
ax.set_ylabel("y", fontsize = 14)

ax.tick_params(axis='both', labelsize=14)

plt.show()
```

### Plotting a Series of Points with scatter()
1. Plot odd values in *red*.
2. Plot even values in *black*.

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.style.use('seaborn')
fig, ax = plt.subplots()

for i in range(len(x_values)):
    if x_values[i] % 2 == 0:
        ax.scatter(x_values[i], y_values[i], c = 'red', s = 100)
    else:
        ax.scatter(x_values[i], y_values[i], c = (0.0, 0.0, 0.0), s = 100)

ax.set_title("Multiple Points", fontsize = 24)
ax.set_xlabel("x", fontsize = 14)
ax.set_ylabel("y", fontsize = 14)

ax.tick_params(axis='both', labelsize=14)
plt.show()
```

### Calculating Data Automatically

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 10)

ax.set_title("Curve", fontsize = 24)
ax.set_xlabel("x", fontsize = 14)
ax.set_ylabel("y", fontsize = 14)
ax.axis([0, 1100, 0, 1100000])

ax.tick_params(axis='both', labelsize=14)
plt.show()
```

### Using a Colormap
1. A ***colormap*** is a series of colors in a gradient that moves from a starting to an ending color.
2. A ***colormap*** is applied to emphasize a pattern in the data. For example, the brighter the smaller values and the darker the larger values.
3. The thrid argugemnt in the `sactter()` function pass the `y_values` to `c`. It colors the points with lower  `y_values` light blue and higher `y_values` dark blue.
4. The forth argugemnt in the `sactter()` function tells which ***colormap*** to use.

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c = y_values, cmap = plt.cm.Blues, s = 10)

ax.set_title("Curve", fontsize = 24)
ax.set_xlabel("x", fontsize = 14)
ax.set_ylabel("y", fontsize = 14)
ax.axis([0, 1100, 0, 1100000])

ax.tick_params(axis='both', labelsize=14)
plt.show()
```

### Saving Your Plots Automatically
1. The first argument si the name of this file.
2. The second argument cut the extra white space around the plot off.
3. To keep the white space around the plot, just omit the second argument.

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c = y_values, cmap = plt.cm.Set1, s = 10)

ax.set_title("Curve", fontsize = 24)
ax.set_xlabel("x", fontsize = 14)
ax.set_ylabel("y", fontsize = 14)
ax.axis([0, 1100, 0, 1100000])

ax.tick_params(axis='both', labelsize=14)
plt.savefig('square_plot1.png', bbox_inches = 'tight')
plt.savefig('square_plot2.png')
```

## Random Walks
### Plotting the Random Walk
1. `-1` for left and `1` for right in `x_direction`.
2. Zero `x_step` value and Non-zero `y_step` value for walking vertically.

```python
from random import choice

class RandomWalks:
    def __init__(self, num_points = 5000):
        self.num_points = num_points
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        while len(self.x_values) < self.num_points: 
            x_direction = choice([-1, 1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance

            y_direction = choice([-1, 1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            if x_step == 0 and y_step == 0:
                continue
            else:
                self.x_values.append(self.x_values[-1] + x_step)
                self.y_values.append(self.y_values[-1] + y_step)

rm = RandomWalks()
rm.fill_walk()
plt.style.use('classic')
fig, ax = plt.subplots()
ax.scatter(rm.x_values, rm.y_values, s = 15)
plt.show()
```

### Styling the Walk
1. The `edgecolor` argument get rid of the black circle around each point.

```python
rm = RandomWalks()
rm.fill_walk()
plt.style.use('classic')

fig, ax1 = plt.subplots()
point_numbers = range(rm.num_points)
ax1.scatter(rm.x_values, rm.y_values, c = point_numbers, cmap = plt.cm.Blues, s = 15)

fig, ax2 = plt.subplots()
point_numbers = range(rm.num_points)
ax2.scatter(rm.x_values, rm.y_values, c = point_numbers, cmap = plt.cm.Blues, edgecolors = 'none', s = 15)

plt.show()
```

### Plotting the Starting and Ending Points
1. Staring point with (x, y) : `rm.x_values[0]` and `rm.y_values[0]`.
2. Ending point with (x, y) : `rm.x_values[1]` and `rm.y_values[-1]`.
### Cleaning Up the Axes
1. `ax.get_xaxis().set_visible(False)`
2. `ax.get_yaxis().set_visible(False)`
### Adding Plot Points
1. Up to 50,000 points, no longer 5000
### Altering the Size to Fill the Screen
1. `fig, ax = plt.subplots(figsize=(15, 9), dpi = 128)`
2. `figsize` takes a touple which is the dimension of the figure window in inches.
3. `dpi` stands for Dots Per Inch, meaning the number of dots can be placed per inch. Larger dpi gives figuers with higher resolution.

```python
rm = RandomWalks(50_000)
rm.fill_walk()

plt.style.use('classic')
fig, ax = plt.subplots(figsize=(15, 9), dpi = 300)

point_numbers = range(rm.num_points)

ax.scatter(rm.x_values, rm.y_values, c = point_numbers, cmap = plt.cm.Blues,edgecolors = 'none', s = 1)
ax.scatter(rm.x_values[0], rm.y_values[0], c = 'green',edgecolors = 'none', s = 100)
ax.scatter(rm.x_values[-1], rm.y_values[-1], c = 'red',edgecolors = 'none', s = 100)

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.show()
```

15-5. Refactoring:

```python
from random import choice

class RandomWalks:
    def __init__(self, num_points = 5000):
        self.num_points = num_points
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        while len(self.x_values) < self.num_points: 
            x_step = self.get_step()

            y_step = self.get_step()
            
            if x_step == 0 and y_step == 0:
                continue
            else:
                self.x_values.append(self.x_values[-1] + x_step)
                self.y_values.append(self.y_values[-1] + y_step)
                
    def get_step(self):
        direction = choice([-1, 1])
        distance = choice([0, 1, 2, 3, 4])
        return direction * distance

rm = RandomWalks()
rm.fill_walk()
plt.style.use('classic')
fig, ax2 = plt.subplots()
point_numbers = range(rm.num_points)
ax2.scatter(rm.x_values, rm.y_values, c = point_numbers, cmap = plt.cm.Blues, edgecolors = 'none', s = 15)

plt.show()
```
